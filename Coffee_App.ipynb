```python
class Coffee:

    def __init__(self, name, price):
        self.name = name
        self.price = price

class Order:

    def __init__(self):
        self.items = []

    def add_item(self, coffee):
        self.items.append(coffee)

        print(f"{coffee.name} was added to the order.")

    def calculate_total(self):
        total = sum(item.price for item in self.items)
        return total

    def show_order(self):
        if not self.items:
            print("No items in the order.")
        else:
            print("Order Items:")
            for item in self.items:
                print(f"{item.name} - ${item.price}")

            print(f"Total: ${self.calculate_total()}\n")

    def checkout(self):
        if not self.items:
            print("No items in the order.")

            return

        self.show_order()

        print("Thank you for your order!")

        confirm = input("Proceed to checkout? (yes/no): ").strip().lower()
        if confirm == "yes":
            print("Order confirmed.")

            self.items = []
            print("Order cleared.\n")
        else:
            print("Order canceled.\n")



    def main():

        menu = [
            Coffee("Espresso", 2.50),
            Coffee("Cappuccino", 3.00),
            Coffee("Latte", 3.50),
            Coffee("Mocha", 4.00),
        ]

        order = Order()

        while True:
            print("\n--- Coffee Shop Menu ---")
            for i, item in enumerate(menu, start=1):
                print(f"{i}. {item.name} - ${item.price}")
            print("5, View Order")
            print("6, Checkout")
            print("7, Exit")

            choice = input("Enter your choice: ")
            if choice in ['1', '2', '3', '4']:
                order.add_item(menu[int(choice) - 1])
            elif choice == '5':
                order.show_order()
            elif choice == '6':
                order.checkout()

            elif choice == '7':
                print("THANKS FOR VISITING, GOODBYE!")

                break

            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    Order.main()
```
